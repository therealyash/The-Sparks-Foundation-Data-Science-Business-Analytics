### Profit vs Discount

```python
fig,ax=plt.subplots(figsize=(20,8))
ax.scatter(df['Sales'],df['Profit'])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show()
```

```python
sns.lineplot(x='Discount',y='Profit',label='Profit',data=df)
plt.legend()
plt.show()
```

### Observation 4

There is no correlation between profit and discount

### Profit vs Quantity

```python
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=df)
plt.legend()
plt.show()
```

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['pink','blue'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

### Observation 5 


The consumer segment yields the highest profit and sales, while the Home Office segment demonstrates comparatively lower figures.

```python
plt.figure(figsize=(12,8))
plt.title('Segment wise Sales in each Region')
sns.barplot(x='Region',y='Sales',data=df,hue='Segment',order=df['Region'].value_counts().index,palette='rocket')
plt.xlabel('Region',fontsize=15)
plt.show()
```

### Observation 6 

Sales are almost similar in every region segmentwise.

```python
df.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['blue','red'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

### Observation 7

The west region exhibits the highest levels of profit and sales, whereas the south region shows comparatively lower figures.

```python
ps = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['blue','orange'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

### Observation 8 

California and New York experience high profitability, while Texas, Pennsylvania, and Ohio incur losses.

```python
t_states = df['State'].value_counts().nlargest(10)
t_states
```

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['yellow','purple'],alpha=0.9,figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

### Observation 9 

#### As a business manager, what are the areas of improvement that can be targeted to increase profitability?

Technology and office supplies generate significant profits, whereas furniture yields lower profitability.

```python
ps = df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['red','lightblue'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

### Observation 10

Phones sub-category demonstrates strong sales performance, while chairs exhibit high sales but lower profitability compared to phones. However, tables and bookmarks sub-categories experience significant losses.
