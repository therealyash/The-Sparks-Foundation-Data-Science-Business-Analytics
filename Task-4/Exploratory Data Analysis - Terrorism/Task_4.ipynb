### US Terror Attacks and Death Injuries.

```python
df.nkillus.plot(kind = 'line', color = 'red', label = "The US's cumulative loss of lives due to confirmed fatalities", linewidth = 2, alpha = 0.8, grid = True, 
                 linestyle = ':', figsize = (20,20), fontsize=15)
df.nwoundus.plot(color = "green", label = "The total count of documented non-fatal injuries in the US", linewidth = 2, alpha = 0.8, grid = True, 
                 linestyle = '-.', figsize = (20,20), fontsize=15)

plt.legend(loc='upper right')     
plt.xlabel('Database Catalog', fontsize=10)              
plt.ylabel('Fatalities and Injuries Count', fontsize=15)

plt.title('The combined tally of confirmed fatalities and non-fatal injuries in the US')            
plt.show()
```

By analyzing the chronological order of the data, it appears that occurrences of attacks on US citizens are infrequent within a specific date range. However, the subsequent year after this uncommon period exhibits a noticeable surge in terrorist acts targeting US citizens. Pinpointing the onset of this increase allows for the identification of influential factors by considering the country's transformations and advancements following that particular date.

### Terrorist Incidents in a Specific Year and their Geographic Location


Exploring Global Terrorist Attacks within a Specific Year

```python
filterYear = df['iyear'] == 1970
```

```python
filterData = df[filterYear] 
reqFilterData = filterData.loc[:,'city':'longitude'] 
reqFilterData = reqFilterData.dropna() 
reqFilterDataList = reqFilterData.values.tolist()
```

```python
map = folium.Map(location = [0, 30], tiles='CartoDB positron', zoom_start=2)
markerCluster = folium.plugins.MarkerCluster().add_to(map)
for point in range(0, len(reqFilterDataList)):
    folium.Marker(location=[reqFilterDataList[point][1],reqFilterDataList[point][2]], popup = reqFilterDataList[point][0]).add_to(markerCluster)
map
```

In 1970, the American continent witnessed 84% of all terrorist attacks, while the Middle East and North Africa, despite being the present-day epicenter of conflicts and terrorism, experienced only a single incident.

### Total Fatality Count in Terror Attacks

```python
killData = df.loc[:,'nkill']
print('People killed by terror attack:', int(sum(killData.dropna())))# drop the NaN values
```

Deaths by terrrorism - 411868

```python
attackData = df.loc[:,'attacktype1':'attacktype1_txt']
typeKillData = pd.concat([attackData, killData], axis=1)
```

```python
typeKillFormatData = typeKillData.pivot_table(columns='attacktype1_txt', values='nkill', aggfunc='sum')
typeKillFormatData
```

```python
typeKillFormatData.info()
```

```python
labels = typeKillFormatData.columns.tolist()
transpoze = typeKillFormatData.T
values = transpoze.values.tolist()
values = [val[0] for val in values]  # Extracting the first element from each sublist

fig, ax = plt.subplots(figsize=(20, 20), subplot_kw=dict(aspect="equal"))
plt.pie(values, startangle=90, autopct='%.2f%%')
plt.title('Types of terrorist attacks that cause deaths')
plt.legend(labels, loc='upper right', bbox_to_anchor=(1.3, 0.9), fontsize=15)
plt.show()
```


The combination of armed assault and bombing/explosion accounts for 77% of fatalities in these attacks, highlighting their frequent utilization in terrorist activities. This significant percentage underscores the grave threat posed by weapons and explosives worldwide.

#### Number of Killed in Terrorist Attacks by Countries.

```python
countryData = df.loc[:,'country':'country_txt']

countryKillData = pd.concat([countryData, killData], axis=1)
```

```python
countryKillFormatData = countryKillData.pivot_table(columns='country_txt', values='nkill', aggfunc='sum')
countryKillFormatData
```

```python
countryKillFormatData.info()
```

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0]=25
fig_size[1]=25
plt.rcParams["figure.figsize"] = fig_size
```

```python
labels = countryKillFormatData.columns.tolist()
labels = labels[:50] #50 bar provides nice view
index = np.arange(len(labels))
transpoze = countryKillFormatData.T
values = transpoze.values.tolist()
values = values[:50]
values = [int(i[0]) for i in values] 
colors = ['red', 'green', 'blue', 'purple', 'yellow', 'brown', 'black', 'gray', 'magenta', 'orange'] 
fig, ax = plt.subplots(1, 1)
ax.yaxis.grid(True)
fig_size = plt.rcParams["figure.figsize"]
fig_size[0]=100
fig_size[1]=100
plt.rcParams["figure.figsize"] = fig_size
plt.bar(index, values, color = colors, width = 0.9)
plt.ylabel('Killed People', fontsize=15)
plt.xticks(index, labels, fontsize=12, rotation=90)
plt.title('Number of people killed by countries')
plt.show()
```
